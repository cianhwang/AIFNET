```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
video = np.zeros((200, 20, 64, 64, 3), dtype=np.uint8)
for i in range(200):
    for j in range(20):
        video[i, j] = cv2.imread('bg_database/'+str(i+1)+'_{:03d}.jpg'.format(j+1))
```

```python
plt.imshow(video[4, 19])
```

```python
np.save('video.npy', video)
```
